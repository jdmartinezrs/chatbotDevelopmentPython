```python
frases = [
    "hola", "buenas", "qu√© tal",
    "necesito ayuda", "puedes ayudarme",
    "gracias", "mil gracias",
    "adi√≥s", "hasta luego", "nos vemos",
    "mantenimiento cr√≠tico", "mantenimiento preventivo", "alerta de reparaci√≥n",
    'proximo mantenimiento',
    '5000','9000','15000',

]

intenciones = [
    "saludo", "saludo", "saludo",
    "ayuda", "ayuda",
    "agradecimiento", "agradecimiento",
    "despedida", "despedida", "despedida",
    "critico", "prevencion", "alerta",
    'mantenimiento',
    'mantenimiento preventivo','mantenimiento reparaci√≥n','mantenimiento cr√≠tico'

]
```

```python
from sklearn.feature_extraction.text import CountVectorizer  # Importa el vectorizador de texto

vectorizador = CountVectorizer()  # Crea un objeto vectorizador
X = vectorizador.fit_transform(frases)  # Aplica vectorizaci√≥n a las frases (cada palabra se convierte en un n√∫mero)
print(X)
```

```python
from sklearn.naive_bayes import MultinomialNB  # Importa el clasificador Naive Bayes

modelo = MultinomialNB()  # Crea un modelo vac√≠o
modelo.fit(X, intenciones)  # Entrena el modelo con las frases vectorizadas y sus etiquetas de intenci√≥n
```

```python
entrada = input("T√∫: ")  # El usuario escribe una nueva frase
entrada_vector = vectorizador.transform([entrada])  # Transforma esa frase a vectores (usando el mismo vocabulario)
intencion = modelo.predict(entrada_vector)[0]  # El modelo predice la intenci√≥n (devuelve una etiqueta)
print("Intenci√≥n detectada:", intencion)  # Muestra la intenci√≥n detectada
```
